Arrays

```python
## Contains Duplicate Elements

class Solution:
    def hasDuplicate(self, nums: List[int]) -> bool:
        freq = {}
        for num in nums:
            if num in freq.keys():
                return True
            else:
                freq[num] = 1
        return False
```

```python
## 2 sum problem

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        freq = {}
        for i in range(len(nums)):
            if target - nums[i] in freq.keys():
                return [freq[target-nums[i]],i]
            freq[nums[i]] = i
        return False
```

Strings

```python
## Check if 2 strings are anagrams of each other

class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False
        freq_s = {}
        freq_t = {}
        for i in range(len(s)):
            freq_s[s[i]] = freq_s.get(s[i],0) + 1
            freq_t[t[i]] = freq_t.get(t[i],0) + 1
        return freq_s == freq_t
```
